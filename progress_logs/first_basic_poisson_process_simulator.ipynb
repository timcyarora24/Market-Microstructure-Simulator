```python
# Creating first simple simulator
#    - Build a basic Poisson process
#    - Generate exponential inter-arrival times
#    - Plote a simple timeline of trade arrivals
#    - Goal: Understand the fundamental randomness
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import default_rng
import yfinance as yf
rng=default_rng()
```

```python
def simulate_trade_arrival(arrival_rate , time_period_minutes):
    #  converting into hours
    time_period_hour=time_period_minutes/60                          
    

    trade_time=[]
    current_time=0
    while current_time<=time_period_hour:
        inter_arrival_time=rng.exponential(1/arrival_rate)
        current_time+=inter_arrival_time
        if current_time <=time_period_hour:
            trade_time.append(current_time)

    return trade_time
```

```python

# Simulate hourly blocks from 9:30 AM to 4:00 PM (13 blocks × 60 minutes)
market_open = 9 * 60 + 30  # 570
market_close = 16 * 60     # 960

hour_blocks = [(t, t + 60) for t in range(market_open, market_close, 60)]
print(hour_blocks)
all_trade_times = []

for i, (start, end) in enumerate(hour_blocks):
    lambda_rate = rng.integers(10, 100)
    print(f'Hour {i+1} | λ: {lambda_rate}')
    
    trades = simulate_trade_arrival(lambda_rate, 60)
    shifted_trades = [start + t*60 for t in trades]
    print("shifted_trades" , shifted_trades)
    print(f'Total trades this hour: {len(shifted_trades)}')
    all_trade_times.extend(shifted_trades)
```
